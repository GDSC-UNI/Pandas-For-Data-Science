# Métodos y atributos de los DataFrames 

En Data Science se suele trabajar con datasets muy grandes, por ello lo más conveniente para visualizar con qué datos nos encontramos trabajando es utilizar el método head que nos muestra las primeras filas del DataFrame.
 
<code>DataFrame.head(n=5)</code>
 
Como vemos en la sintaxis, este método tiene un parámetro n que indica cuántas filas vamos a querer ver, por defecto, este valor es 5, sin embargo, podemos modificarlo y seleccionar cuántas filas queremos ver.




```python
df_titanic.head(10)
```

Así mismo también se puede utilizar el método tail para visualizar las n últimas filas.

```python
df_titanic.tail(10)
```

Hay veces en que no visualizamos datos relevantes ni en las primeras filas ni en las últimas filas, por ello existe también el método sample que nos permite visualizar filas elegidas de manera aleatoria y poder hacer algunas hipótesis acerca de nuestros datos.

```python
df_titanic.sample(10)
```

Para conocer el tamaño del dataset con el que nos encontramos trabajando usamos el atributo shape que nos da como resultado una tupla con el número de filas y columnas de nuestro dataset.

```python
df_titanic.shape
```

Al realizar operaciones matemáticas con los DataFrame para obtener medidas de estadística como la media, desviación estándar, rango intercuartil, etc. A veces, el resultado de estas operaciones tiene números flotantes con muchos decimales que dificultan una buena visualización de nuestro DataFrame, para solucionar esto, especificamos el formato con el que queremos que pandas nos muestre los números flotantes.

```python
pd.options.display.float_format='{:,.3f}'.format
```

Anteriormente mencionamos operaciones matemáticas clásicas en estadística, como las medidas de tendencia central u otras métricas. Los DataFrames contienen un método que nos ayuda a obtener estos valores de nuestros datos de manera rápida. Este método es *describe*.

<code>DataFrame.describe(percentiles=None, include=None, exclude=None, datetime_is_numeric=False)<code>


```python
df_titanic.describe()
```

Como vemos en la sintaxis, hay un parámetro llamado include el cual está por defecto como None, al mantenerse así, el método describe solo realiza las medidas de valores tipo numérico y obvia los valores tipo object. Si queremos hacer una descripción de todas las columnas, cambiamos el valor de este parámetro por 'all'. 

```python
df_titanic.describe(include='all')
```

Otro método de los DataFrame que nos es de mucha utilidad al momento de hacer el análisis de nuestros datos es *info*, el cual nos da una información de las features con su cantidad de datos no nulos y el tipo de datos de estos.
 
<code> DataFrame.info(verbose=None, buf=None, max_cols=None, memory_usage=None, show_counts=None, null_counts=None)</code>
 
Generalmente con este método no realizan modificaciones en sus parámetros, salvo de ocasiones muy especiales, revisar la <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html#pandas.DataFrame.info">documentación</a> si desea profundizar más en este método.


```python
df_titanic.info()
```

Para solo conocer los tipos de datos de las features con las que estamos trabajando, utilizamos el atributo *dtypes* 

```python
df_titanic.dtypes
```

Como podemos apreciar, vemos que hay features, cómo Name, Sex, Age, Ticket, Cabin y Embarked que se encuentran identificadas como tipo objeto, para convertir estos tipos objetos a string, utilizamos el método.
 
<code>DataFrame.convert_dtypes(infer_objects=True, convert_string=True, convert_integer=True, convert_bool=True, convert_floating=True)</code>
 
Al utilizar este método, con sus parámetros por defecto, las columnas serán convertidas al mejor tipo de dato posible, para nuestro caso, convertirá los objects a string.



```python
df_titanic.convert_dtypes().dtypes
```

Esta conversión no es la más correcta ya que existen datos que pueden ser guardados como tipos de datos categóricos, para esto tenemos que realizar un análisis de los valores únicos de cada feature, esto lo podemos realizar con el método:

<code> DataFrame.unique(axis=0, dropna=True)</code>

```python
df_titanic.nunique()
```

Podemos ver del resultado que survived y sex tienen 2 valores únicos; Embarket y Pclass, 3. Por ende, convertimos los datos de estas features en tipo categorial utilizando el siguiente método:

<code>DataFrame.astype(dtype, copy=True, errors='raise') </code>


```python
df_titanic[['Survived', 'Pclass', 'Sex', 'Embarked']] = df_titanic[['Survived', 'Pclass', 'Sex', 'Embarked']].astype('category')
```

```python
df_titanic['Sex'].unique()
```
